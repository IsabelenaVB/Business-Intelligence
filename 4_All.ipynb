#**NAIVE BAYES**

```python
import numpy as np
import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix

# Cargar el archivo Excel
nxl = "/content/drive/MyDrive/SEMESTRE-6/4. BD2_NBA_Data_M_Int.xlsx"
XDB = pd.read_excel(nxl, sheet_name=0)

# Seleccionar las variables
XDB = XDB[["salary", "games_played_0y1", "minutes_per_game"]].dropna()

# Separar las variables de entrada y salida
XD = XDB[["salary", "minutes_per_game"]]
yd = XDB["games_played_0y1"]

# Modelo Naive Bayes
mnb = GaussianNB()
mnb.fit(XD, yd)

# Obtener información del modelo
u = mnb.theta_
print("La media de las variables es:", np.round(u, 2))
var = mnb.var_
print("La varianza de las variables es:", np.round(var, 2))
sigma = np.sqrt(var)
print("La desviación estándar es:", np.round(sigma, 2))

# Determinar los límites
LS = u + sigma
LI = u - sigma
print("El límite superior es:", np.round(LS, 2))
print("El límite inferior es:", np.round(LI, 2))


# Crear la matriz de confusión
y_pred = mnb.predict(XD)
cm = confusion_matrix(yd, y_pred)
print("La matriz de confusión es:\n", cm)

# Métricas de desempeño
exactitud = (cm[0,0] + cm[1,1]) / cm.sum()
precision = cm[1,1] / (cm[1,1] + cm[0,1]) if (cm[1,1] + cm[0,1]) != 0 else 0
sensibilidad = cm[1,1] / (cm[1,1] + cm[1,0]) if (cm[1,1] + cm[1,0]) != 0 else 0
especificidad = cm[0,0] / (cm[0,0] + cm[0,1]) if (cm[0,0] + cm[0,1]) != 0 else 0

print("La exactitud es:", exactitud)
print("La precisión es:", precision)
print("La sensibilidad es:", sensibilidad)
print("La especificidad es:", especificidad)
```

#**ANALISIS DE RESULTADOS**
Para este analisis utilizamos tres modelos que me permitieran determinar si el número de juegos jugados por jugador de la NBA está por encima del promedio. Se utilizaron datos de la NBA con variables como salary (salario del jugador), age (edad del jugador), height (altura), weight (peso), games_started (juegos iniciados), minutes_per_game (promedio de minutos jugados por partido), assists (promedio de asistencias por partido), blocks (promedio de bloqueos por partido) y steals (promedio de robos de balón por partido). e implementaron tres modelos: Naive Bayes, Árbol de Decisión y Clustering (K-Means).

El modelo Naive Bayes mostró un desempeño aceptable, con una exactitud del 70.3%. Sin embargo, su especificidad del 91.8% refleja que el modelo clasifica con mayor certeza a los jugadores con menor participación.

El Árbol de Decisión fue el modelo más preciso, logrando una exactitud del 95.5%. Este resultado indica que el modelo es altamente eficiente para clasificar correctamente a los jugadores en función de las variables de entrada. Su sensibilidad del 93.7% y especificidad del 96.8% confirman su balance en la identificación de ambas categorías. Además, la precisión del modelo fue del 95.7%, reflejando un excelente desempeño en la predicción de los jugadores con alta participación en los partidos.

Por último, el modelo K-Means permitió segmentar a los jugadores en 5 clusters con base en su rendimiento y características socioeconómicas. La exactitud del modelo fue del 71.2%, con una sensibilidad del 47.3% y una especificidad del 89%. Esto sugiere que el modelo es más efectivo al identificar jugadores con menor tiempo de juego, pero tiene dificultades para clasificar correctamente a los jugadores con mayor participación.
